### Create Data Frame and Table

```
# File location and type
#file_location = "/FileStore/tables/AB_NYC_2019-7.csv"
file_location = "/FileStore/tables/AB_NYC_2019_no_name.csv"
file_type = "csv"
 
# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","
 
# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("schema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)
 
# Convert the data types from string to integer
from pyspark.sql.functions import col
from pyspark.sql.types import StringType,BooleanType,DateType,IntegerType,FloatType
df = df.withColumn("price",col("price").cast(IntegerType())) \
     .withColumn("minimum_nights",col("minimum_nights").cast(IntegerType())) \
     .withColumn("number_of_reviews",col("number_of_reviews").cast(IntegerType())) \
     .withColumn("reviews_per_month",col("reviews_per_month").cast(IntegerType())) \
     .withColumn("calculated_host_listings_count",col("calculated_host_listings_count").cast(IntegerType())) \
     .withColumn("availability_365",col("availability_365").cast(IntegerType()))
  
# display(df)

df.printSchema()
df.show(truncate=False)
```

```
# Create a view or temporary table
temp_table_name = "AB_NYC_2019_csv"
 
df.createOrReplaceTempView(temp_table_name)
```

```
# Import the required libraries
 
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler,StringIndexer
from pyspark.ml import Pipeline
```

```
# Use StringIndexer to convert the categorical columns to hold numerical data
 
neighbourhood_group_indexer = StringIndexer(inputCol='neighbourhood_group',outputCol='neighbourhood_group_index',handleInvalid='keep')
```

```
# Vector assembler to create vector of input features
#featureassembler = VectorAssembler(inputCols=["reviews_per_month","minimum_nights","number_of_reviews","calculated_host_listings_count","availability_365","neighbourhood_group_index"], outputCol="Independent")
featureassembler = VectorAssembler(inputCols=["minimum_nights","number_of_reviews","calculated_host_listings_count","availability_365","neighbourhood_group_index"], outputCol="Independent")
```

```
# Define the pipeline
pipe = Pipeline(stages=[neighbourhood_group_indexer,featureassembler])
```

```
# Create a 70-30 train/test split
train_data, test_data = df2.randomSplit(weights=[0.70,0.30], seed=42)
```

```
fitted_pipe=pipe.fit(train_data)
```

```
#train_data.show(100)
display(train_data)
```

```
# Store the results in a data frame
train_data=fitted_pipe.transform(train_data)
#train_data.show(100)
display(train_data)
```

```
# Initializing a Linear Regression model
#lr_model = LinearRegression(maxIter = 10, regParam = 0.3, elasticNetParam = 0.8, featuresCol='Independent', labelCol='price')
lr_model = LinearRegression(featuresCol='Independent',labelCol='price')
```

```
# Fit the model on the train data
fit_model = lr_model.fit(train_data.select(['Independent','price']))
```

```
# Transform the test data using the model to predict price
test_data=fitted_pipe.transform(test_data)
#test_data.show(100)
display(test_data)
```

```
# Store the results in a dataframe
results = fit_model.transform(test_data)
display(results)
#results.show(100)
```

```
#results.select(['price', 'prediction']).show
```

```
test_results = fit_model.evaluate(test_data)
test_results.residuals.show()
```

```
print('MAE for test set:', test_results.meanAbsoluteError)
print('r2 for test set:', test_results.r2)
print('RMSE for test set:', test_results.rootMeanSquaredError)
```
